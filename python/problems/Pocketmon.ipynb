```python
import random

class Pikachu:
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15,20)
        self.exp = 0
        
    def bark(self):
        print('pikachu!!!!!!!')
        
    def attack(self, yours):
        skill = input(f'{self.name}은 어떤 공격을 할까요? (몸통박치기, 천만볼트, 항복)\n===========================>')
        if skill == '몸통박치기':
            return self.body_attack(yours)
        elif skill == '천만볼트':
            return self.thousond_volt(yours)
        elif skill == '항복':
            return '항복'
        else:
            print('스킬명을 잘못 입력했습니다. 너의 턴은 날아갔다')
            
       
    def body_attack(self, yours):
        print(f'{self.name}가 {yours.name}에게 몸통박치기 공격을 했습니다 ')
        if random.randint(0, 2):
            yours.hp -= self.level * 5
            print(f'{yours.name}의 hp가 {self.level * 5}만큼 줄었습니다')
            return self.attack_result(yours)        
        else:
            print('공격에 실패했다')   
        
        
    def thousond_volt(self, yours):
        print(f'{self.name}가 {yours.name}에게 천만볼트 공격을 했습니다 ')
        if not random.randint(0, 2):
            yours.hp -= self.level * 7
            print(f'{yours.name}의 hp가 {self.level * 7}만큼 줄었습니다')
            return self.attack_result(yours)
        else:
            print('공격에 실패했다')        
    

    def attack_result(self, yours):
        if yours.hp <= 0:
            print(f'{self.name}가 승리하였습니다!')
            self.exp += yours.level * 30
            yours.hp = yours.level * random.randint(15,20)     # 상대방 체력 0이 되어 회복
            return self.level_up()
        else:
            print(f'{self.name} : {self.hp}         {yours.name} : {yours.hp}')
        
    
    
    def level_up(self):
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp -= self.level * 100
            print(f'{self.name}가 {self.level}로 레벨업을 하였습니다')
        else:
            print(f'{self.name}의 경험치가 {self.exp}가 되었습니다')
        self.hp = self.level * random.randint(15,20)          # 경기 결과에 따라 내 hp도 회복
        
        print('모든 hp를 회복시켰습니다. 계속 싸우세요')
        

        
```
